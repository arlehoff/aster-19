import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

create a numpy array x ranging from 0 to 1 in 100 steps 

```python
x = np.linspace(0, 1, 100)
```

define functions that return sin(x) and cos(x)

```python
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
```

use the functions to set y values 

```python
y_sin = sin_function(x)
y_cos = cos_function(x)
```

create a multipanel plot (1 row, 2 columns)

```python
fig, axes = plt.subplots(1, 2, figsize= (10, 4))
```

plot sin(x) vs. x in the left panel 

```python
axes[0].plot(x, y_sin)
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
```

plot cos(x) vs. x in the right panel 

```python
axes[1].plot(x, y_cos)
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('sin(x)')
```

save the figure as pdf 

```python
plt.savefig('sine_cosine_plot.pdf')
plt.show()
```

create 1000 random numbers uniformly distrubted between [0,1]

```python
random_numbers = np.random.uniform(0, 1, 100)
```

create a histogram of the random numbers into 100 bins 

```python
plt.figure(figsize=(8,6))
plt.hist(random_numbers, bins=100, edgecolor='black') 
plt.xlabel('Value of Uniform Random Numbers')
plt.ylabel('Frequency of Uniform Random Numbers')
plt.title('Histogram of Uniform Random Numbers')
```

Save the histogram figure as a pdf 

```python
plt.savefig('uniform_random_histogram.pdf')
plt.show()
```

Generate 100 random numbers from a logistic distribution 

```python
logistic_data = np.random.logistic(loc=0.0, scale=1.0, size=1000)
```

create a histogram of the logistic random numbers into 100 bins 

```python
plt.figure(figsize=(8, 6))
plt.hist(logistic_data, bins=100, edgecolor='black')
plt.xlabel('Value of Logistic Random Numbers')
plt.ylabel('Frequency of Logistic Random Numbers')
plt.title('Histogram of Logistic Random Numbers')
```

save the histogram figure as a PDF

```python
plt.savefig('logistic_random_histogram.pdf')
plt.show()
```
