import libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

create an array x ranging from 0 to 1 in 100 steps 

```python
x = np.linspace(0, 1, 100)
```

define functions that return sin(x) and cos(x)

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

use functions to set y values 

```python
y_sin = sin_function(x)
y_cos = cos_function(x)
```

create a multipanel plot (1 row, 2 columns) 

```python
fig, axes = plt.subplots(1, 2, figsize= (10, 4))
```

plot sin(x) vs. x in the left panel 

```python
axes[0].plot(x, y_sin)
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
```

plot cos(x) vs. x in the right panel 

```python
axes[1].plot(x, y_cos)
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
```

adjust the layout 

```python
plt.tight_layout()
```

save figure as a pdf

```python
plt.savefig('sine_cosine_plot.pdf')
plt.show()
```

Generate 1000 random numbers uniformly distributed between [0,1]

```python
random_numbers = np.random.uniform(0, 1, 100)
```

create a historgram of the random numbers into 100 bins

```python
plt.figure(figsize=(8,6))
plt.hist(random_numbers, bins=100, edgecolor='black')
plt.xlabel('Value of Uniform Random Numbers')
plt.ylabel('Frequency of Uniform Random Numbers')
plt.title('Histogram of Uniform Random Numbers')
```

save the histogram figure as PDF

```python
plt.savefig('uniform_random_histogram.pdf')
plt.show()
```
